```python
#Display GDP

gdp = pd.read_excel('gdp_data.xlsx', sheetname='Total GDP')
gdp.head()
```

```python
#Display Military

military = pd.read_excel('military_data.xlsx', sheetname='Total Military')
military.head()
```

```python
#Sort top 10 countries for military spending

top_military= military.sort_values(by='2006_budget', ascending= False) 
topmilitary = top_military.head(10)
topmilitary
```

```python
# Military Budget vs GDP Comparison across 5 years.
# For the purpose of comparision, a ratio of budget spent on military and GDP is taken for each year.
# An interactive stacked Plotly bar graph is created for this purpose.

military_gdp = pd.merge(topmilitary, gdp, on='Country Name', how='inner')
for year in range(2006, 2016):
    military_gdp[str(year)+'_military_gdp']=military_gdp[str(year)+'_budget']*100/military_gdp[str(year)+'_gdp']
```

```python
trace1 = go.Bar(
    x=military_gdp['Country Name'],
    y=military_gdp['2011_military_gdp'],
    name='2011'
)
trace2 = go.Bar(
    x=military_gdp['Country Name'],
    y=military_gdp['2012_military_gdp'],
    name='2012'
)
trace3 = go.Bar(
    x=military_gdp['Country Name'],
    y=military_gdp['2013_military_gdp'],
    name='2013'
)
trace4 = go.Bar(
    x=military_gdp['Country Name'],
    y=military_gdp['2014_military_gdp'],
    name='2014'
)
trace5 = go.Bar(
    x=military_gdp['Country Name'],
    y=military_gdp['2015_military_gdp'],
    name='2015'
)
data = [trace1, trace2, trace3, trace4, trace5]
layout = go.Layout(
    barmode='stack'
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='stacked-bar')
```

```python
#Display military speinding country wise from year 2006-2015

militarytransposed = topmilitary.set_index(['Country Name']).transpose()
militaryplot = militarytransposed.plot.line(legend=True, 
                                            figsize=(20,10),
                                            title='Military Spending 2006-2015')
militaryplot.set_ylabel('Military Rate %')
militaryplot.set_xlabel('Year')
plt.show()
```

```python
#Display GDP Per capita

gdppc = pd.read_excel('gdp_data.xlsx', sheetname='GDP Per Capita')
countrylist = ['United States', 'United Kingdom', 'China', 'France', 'Japan', 'Russia', 'Italy', 'Saudi Arabia', 'South Korea']
topc = gdppc[gdppc['Country'].isin(countrylist)]
topc
```

```python
#display military per capita

militarypc = pd.read_excel('military_data.xlsx', sheetname='Military Per Capita')
countrylist = ['United States', 'United Kingdom', 'China', 'France', 'Japan', 'Russia', 'Italy', 'Saudi Arabia', 'South Korea']
topm = militarypc[militarypc['Country'].isin(countrylist)]
topm
```

```python
#merge

militarypc_gdppc = pd.merge(topm, topc, on='Country', how='left')
for year in range(2006, 2016):
    militarypc_gdppc[str(year)+'_militarypc_gdppc']=militarypc_gdppc[str(year)+'_militarypc']*100/militarypc_gdppc[str(year)+'_gdppc']
militarypc_gdppc
```

```python
#Plot graph

trace1 = go.Scatter(
    x=militarypc_gdppc['Country'],
    y=militarypc_gdppc['2006_militarypc_gdppc'],
    name='2006'
)
trace2 = go.Scatter(
    x=militarypc_gdppc['Country'],
    y=militarypc_gdppc['2007_militarypc_gdppc'],
    name='2007'
)
trace3 = go.Scatter(
    x=militarypc_gdppc['Country'],
    y=militarypc_gdppc['2008_militarypc_gdppc'],
    name='2008'
)
trace4 = go.Scatter(
    x=militarypc_gdppc['Country'],
    y=militarypc_gdppc['2009_militarypc_gdppc'],
    name='2009'
)
trace5 = go.Scatter(
    x=militarypc_gdppc['Country'],
    y=militarypc_gdppc['2010_militarypc_gdppc'],
    name='2010'
)
data = [trace1, trace2, trace3, trace4, trace5]
layout = go.Layout(
    barmode='Line'
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Line-Plot')
```

```python
#Fastest growing countries in military spending in percentage and value

military['Growth Rate percentage'] = ((military['2015_budget']-military['2006_budget'])/military['2006_budget'])*100
military['Growth Rate value'] = military['2015_budget']-military['2006_budget']
military.head(10)
```

```python
#Sort top Countries in military spending in percentage and in values
fastest_percentage = military.sort_values(by='Growth Rate percentage', ascending= False).head(10)
fastest_value = military.sort_values(by='Growth Rate value', ascending= False).head(10)
```

```python
#Plot military growth in percentage
data = [go.Histogram2d(
    x=fastest_percentage['Country Name'],
    y=fastest_percentage['Growth Rate percentage']
)]
py.iplot(data, filename='2D Histogram')
    
```

```python
#Plot military growth in Values
data1 = [go.Bar(
            x=fastest_value['Growth Rate value'],
            y=fastest_value['Country Name'],
            orientation = 'h'
)]

py.iplot(data1, filename='horizontal-bar')
    
```
